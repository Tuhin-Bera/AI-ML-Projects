```python
ratings = pd.read_csv(r"BX-Book-Ratings.csv", sep=";", on_bad_lines="skip", encoding="latin-1")
```

```python
ratings.rename(columns={"User-ID":"user_id", "Book-Rating":"rating"}, inplace=True)
```

```python
ratings["user_id"].value_counts()
```

```python
ratings["user_id"].value_counts().shape     ## these are the user who gives ratings
```

```python
x = ratings["user_id"].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings["user_id"].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head(3)
```

```python
## now we are joing rating and books df together
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head(3)
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
number_rating.rename(columns={"rating":"num_of_rating"}, inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating, on="title")
```

```python
final_rating
```

```python
final_rating["num_of_rating"] >= 50
```

```python
final_rating = final_rating[final_rating["num_of_rating"] >= 50]   ## this the df where user give more than 200 ratings and the books which have 50 or more ratings
```

```python
final_rating
```

```python
final_rating.drop_duplicates(["user_id", "title"], inplace=True)
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns="user_id", index="title", values="rating")   #  we are matrics format pivot table 
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(algorithm="brute")
nn.fit(book_sparse)
```

```python
distances, suggestions = nn.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
book_pivot.index
```

```python
book_pivot.index[237]
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
np.where(book_pivot.index == "Animal Farm")[0][0]     ## finding the  index of the book
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances2, suggestions2 = nn.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    for i in range(len(suggestions2)):
        if i == 0:
            print("The suggestion book", book_name, "are :")
        if not i:
             print(book_pivot.index[suggestions2[i]])           ## building a functio to find the recom books
```

```python
recommend_book("Animal Farm")
```
